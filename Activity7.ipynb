<h1>CASUALITY ANALYSIS</h1>
<h2>TASK</h2>

```python
casual = (
    df.groupby('targtype1_txt')
      .agg(killed      =('nkill','sum'),
           wounded     =('nwound','sum'),
          incidents   =('iyear','size'))
      .assign(average_killed_per_incident=lambda x: (x['killed'] / x['incidents']).round(2))
      .sort_values('killed', ascending=False))
casual.head(10)
```

<h1> Which targets the most severity</h1>

```python
casual['incidents'].head(10).plot.bar(title='Top 10 average killed per incident')
```

```python
targets = ['Private Citizens & Property', 'Military', 
           'Police', 'Government (General)', 'Business',
           'Transportation', 'Religious Figures/Institutions',
           'Terrorists/Non-State Militia', 'Airports & Aircraft']

for target in targets:
    sub = df[df['targtype1_txt'] == target].copy()
    
    # Drop rows with NaN in latitude or longitude
    sub = sub.dropna(subset=['latitude', 'longitude'])
    
    if sub.empty:
        print(f"⚠️ No data for target: {target}")
        continue
    
    # Create map centered on median location
    m_target = folium.Map(
        location=[sub['latitude'].median(), sub['longitude'].median()],
        tiles='OpenStreetMap',
        zoom_start=3
    )
    
    # Add heat layer
    HeatMap(
        data=sub[['latitude', 'longitude']].values.tolist(),
        radius=14,
        blur=20,
        min_opacity=0.3
    ).add_to(m_target)
    
    # Save the map to an HTML file
    fname = f"heat_target_{target.split()[0]}.html"
    m_target.save(fname)
    print(f"✅ saved {fname}  ({len(sub):,} points)")
```

<h1>NATIONALITY ANALYSIS</h1>
<h2>TASK</h2>

```python
nationality = (
    df.dropna(subset=['country_txt'])
      .groupby('country_txt')
      .agg(killed   =('nkill','sum'),
           wounded  =('nwound','sum'),
          incidents=('iyear','size'))
      .sort_values('incidents', ascending=False))
nationality.head(15)
```

<p>TOP 10 HIGHEST COUNTRIES SINCE 2005 </p>

```python
nationality['incidents'].head(5).plot.bar(title='Top 5 since 2005')
```

```python
countries = ['Iraq', 'Pakistan', 'Afghanistan', 'India', 'Colombia']

for c in countries:
    sub = df[df['country_txt'] == c].copy()
    
    # Drop rows with missing latitude or longitude
    sub = sub.dropna(subset=['latitude', 'longitude'])
    
    if sub.empty:
        print(f"⚠️ No valid data for country: {c}")
        continue
    
    # Create base map centered on median location
    country_map = folium.Map(
        location=[sub['latitude'].median(), sub['longitude'].median()],
        tiles='OpenStreetMap',
        zoom_start=4
    )
    
    # Add heatmap layer
    HeatMap(
        data=sub[['latitude', 'longitude']].values.tolist(),
        radius=12,
        blur=16,
        min_opacity=0.3
    ).add_to(ctry_map)
    
    # Save map to HTML file
    file_name = f"heat_perp_{c}.html"
    country_map.save(file_name)
    print(f"✅ saved {file_name}  ({len(sub):,} points)")
```
